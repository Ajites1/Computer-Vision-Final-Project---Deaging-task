```python
import os
import numpy as np
import glob
from PIL import Image
directory_path = "/projects/standard/csci5561/shared/G8/data/AgeDB" 
```

```python
breakdown = {}
for root, dirs, files in os.walk(directory_path):
    for filename in files:
        path = os.path.join(root, filename)
        img = Image.open(path)
        img_array = np.array(img)
        details = filename.split('_')
        img = Image.open(path)
        image_array = np.array(img)
        img_number = details[0]
        name = details[1]
        age = details[2]
        np.save(f"/projects/standard/csci5561/shared/G8/data/AgeDB_Numpy/{img_number}_{name}_{age}.npy", image_array)
        
        if(name in breakdown):
            breakdown[name][f'{age}_{img_number}'] = f"/projects/standard/csci5561/shared/G8/data/AgeDB_Numpy/{img_number}_{name}_{age}.npy"
        else:
            breakdown[name] = {}
            breakdown[name][f'{age}_{img_number}'] = f"/projects/standard/csci5561/shared/G8/data/AgeDB_Numpy/{img_number}_{name}_{age}.npy"
for key in breakdown:
    print(key)
    for a in breakdown[key]:
        print(f'          {a}: {breakdown[key][a]}')
```

```python
import json
# Save the dictionary to a JSON file
with open('/projects/standard/csci5561/shared/G8/data/AgeDB.json', 'w') as f:
    json.dump(breakdown, f, indent=4)
```

```python
# Load the dictionary back from the file
with open('/projects/standard/csci5561/shared/G8/data/AgeDB.json', 'r') as f:
    loaded_dict = json.load(f)
for i in loaded_dict:
    print(f'{i}')
    for j in loaded_dict[i]:
        print(f'\t {j}: {loaded_dict[i][j]}')
```
