```python
import os
import numpy as np
import glob
from PIL import Image
directory_path = "/projects/standard/csci5561/shared/G8/data/face_age" 
```

```python
breakdown = {}
count = 0
for root, dirs, files in os.walk(directory_path):
    for filename in files:
        path = os.path.join(root, filename)
        if('checkpoint' in path):
            continue
        img = Image.open(path)

        path = path.split('/')
        age = path[-2]

        img_array = np.array(img)


        np.save(f"/projects/standard/csci5561/shared/G8/data/face_age_Numpy/{count}_{age}.npy", img_array)
        
        if(age in breakdown):
            breakdown[age][str(count)] = f"/projects/standard/csci5561/shared/G8/data/face_age_Numpy/{count}_{age}.npy"
        else:
            breakdown[age] = {}
            breakdown[age][str(count)] = f"/projects/standard/csci5561/shared/G8/data/face_age_Numpy/{count}_{age}.npy"

        count+=1
print(count)
```

```python
import json
# Save the dictionary to a JSON file
with open('/projects/standard/csci5561/shared/G8/data/face_age.json', 'w') as f:
    json.dump(breakdown, f, indent=4)
```
