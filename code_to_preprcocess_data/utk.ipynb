```python
import os
import numpy as np
import glob
from PIL import Image
directory_path = "/projects/standard/csci5561/shared/G8/data/utk" 
```

```python
breakdown = {}
count = 0
for root, dirs, files in os.walk(directory_path):
    for filename in files:
        path = os.path.join(root, filename)
        
        img = Image.open(path)
        
        path = path.split('/')
        info = path[-1]
        age = info.split('_')[0]

        img_array = np.array(img)


        np.save(f"/projects/standard/csci5561/shared/G8/data/utk_Numpy/{count}_{age}.npy", img_array)
        
        if(age in breakdown):
            breakdown[age][str(count)] = f"/projects/standard/csci5561/shared/G8/data/utk_Numpy/{count}_{age}.npy"
        else:
            breakdown[age] = {}
            breakdown[age][str(count)] = f"/projects/standard/csci5561/shared/G8/data/utk_Numpy/{count}_{age}.npy"

        count+=1
print(count)
```

```python
import json
# Save the dictionary to a JSON file
with open('/projects/standard/csci5561/shared/G8/data/qtk.json', 'w') as f:
    json.dump(breakdown, f, indent=4)
```
