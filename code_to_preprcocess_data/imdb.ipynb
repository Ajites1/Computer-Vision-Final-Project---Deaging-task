```python
import os
import numpy as np
import glob
from PIL import Image
directory_path = "/projects/standard/csci5561/shared/G8/data/imdb_crop" 
import csv

file_path = '/projects/standard/csci5561/shared/G8/data/full_path.csv'  # Replace with the actual path to your CSV file
file_path1 = '/projects/standard/csci5561/shared/G8/data/ages.csv'  # Replace with the actual path to your CSV file
file_path2 = '/projects/standard/csci5561/shared/G8/data/celeb_id.csv'  # Replace with the actual path to your CSV file

```

```python
breakdown = {}
breakdown2 = {}
with open(file_path, 'r', newline='') as csvfile:
    # Create a CSV reader object
    full_path = csv.reader(csvfile)
    with open(file_path1, 'r', newline='') as csvfile:
        # Create a CSV reader object
        ages = csv.reader(csvfile)
        with open(file_path2, 'r', newline='') as csvfile:
        # Create a CSV reader object
            ids = csv.reader(csvfile)
        
            for row_a in ages:
                break
            for row_path in full_path:
                break
            for row_id in ids:
                break
            for i in range(len(row_a)):
                p = row_path[i]
                filename = p
                age = row_a[i]
                celeb_id = row_id[i]
                _, p = p.split("/")
                p = p.split("_")
                person_id = p[0]
                img_number = p[1]
                birth_year = p[2][0:4]
                photo_year = p[3][0:4]
                if not(os.path.exists(filename)):
                    print(f'{filename} does not exist')
                    path = os.path.join(directory_path , filename)
                    img = Image.open(path)
                    image_array = np.array(img)
                    np.save(f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy", image_array)
                if(age in breakdown):
                    if(celeb_id in breakdown[age]):
                        breakdown[age][celeb_id][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
                    else:
                        breakdown[age][celeb_id] = {}
                        breakdown[age][celeb_id][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
                else:
                    breakdown[age] = {}
                    breakdown[age][celeb_id] = {}
                    breakdown[age][celeb_id][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
            
                
                if(celeb_id in breakdown2):
                    if(age in breakdown2[celeb_id]):
                        breakdown2[celeb_id][age][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
                    else:
                        breakdown2[celeb_id][age] = {}
                        breakdown2[celeb_id][age][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
                else:
                    breakdown2[celeb_id] = {}
                    breakdown2[celeb_id][age] = {}
                    breakdown2[celeb_id][age][img_number] = f"/projects/standard/csci5561/shared/G8/data/imdb_Numpy/{person_id}_{img_number}_{birth_year}_{photo_year}.npy"
        
```

```python
import json
# Save the dictionary to a JSON file
with open('/projects/standard/csci5561/shared/G8/data/fixed_imdb_age_first.json', 'w') as f:
    json.dump(breakdown, f, indent=4)

with open('/projects/standard/csci5561/shared/G8/data/fixed_imdb_person_first.json', 'w') as f:
    json.dump(breakdown2, f, indent=4)
```

```python
# Load the dictionary back from the file
with open('/projects/standard/csci5561/shared/G8/data/fixed_imdb_age_first.json', 'r') as f:
    loaded_dict = json.load(f)
for i in loaded_dict:
    print(f'{i}')
    for j in loaded_dict[i]:
        print(f'\t {j}: {loaded_dict[i][j]}')
```
